```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings ('ignore')
```

```python

df= pd.read_csv('auto_imports.csv',na_values = ['?'],header=0,names=['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinder','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price'])
df
```

```python
plt.rcParams['figure.figsize']=(9,7)
sns.distplot(x= df.price)
plt.show()
```

#### The following price of the automobile data is right skewed  and it has the majority of the price distribution on the centre of the data


```python
make_name= df.make.value_counts().index
make_values= df.make.value_counts().values
plt.pie(x=make_values,labels=make_name,autopct='%1.2f%%')
plt.show()
```

#### Observation
The  top 3 brands of automobiles are from the ' Toyota',' Nissan',' Mazda' with respective to the models produced by them and not related to price. Overall the premium brands are very less comparatively the economic brand based on the demand.


```python
sns.barplot(x=df['make'],y=df['price'])
plt.xticks(rotation=90)
plt.show()
```

#### Observation
The top brands based on the grounds of price are 'Mercedes-Bensz','jaguar','porsche','Bmw'  among the listed brands of cars.
These brands are classified as premium segments and has limited variants for limited customers.

```python
df['body-style'].value_counts()
body_style_name= df['body-style'].value_counts().index
body_style_values= df['body-style'].value_counts().values
plt.pie(x=body_style_values,labels=body_style_name,autopct="%1.2f%%")
plt.show()
```

#### Observation
Based on the body style ,most of the cars are from the 'Sedan'(47%) following 'Hatchback'(34%) and comparatively the 'Çonvertible '(2%) and 'Hardtop'(4%) are very less.It accounts only less than 5% of the total car distribution in this dataset


```python
plt.subplot(1,2,1)
df['fuel-type'].value_counts()
fuel_name= df['fuel-type'].value_counts().index
fuel_values= df['fuel-type'].value_counts().values
plt.pie(x=fuel_values,labels=fuel_name,autopct='%1.2f%%')
plt.show()

plt.subplot(1,2,2)
sns.barplot(x=df['fuel-type'],y=df['price'])
plt.show()
```

#### Observation
The classification of the fuel type cars are divided into disel and petrol in which around 90% cars fall under the disel 
and the rest in petrol .As well as the price of the diesel cars belong to premuim variant and the petrol cars are priced lower
than diesel variants.

```python
sns.barplot(x=df['num-of-cylinder'],y=df['price'])
plt.xlabel('Number of Cyclinder')
plt.ylabel('Price')
plt.show()
```

```python
fig=plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.scatterplot(df["peak-rpm"],df["price"],df["num-of-cylinder"],color="orange")
plt.xticks(rotation=90)
plt.xlabel("peak-rpm",fontsize=15)
plt.ylabel("price",fontsize=15)

plt.subplot(1,3,2)
plt.scatter(df["city-mpg"],df["price"],color="blue")
plt.xticks(rotation=90)
plt.xlabel("city-mpg",fontsize=15)

plt.subplot(1,3,3)
plt.scatter(df["highway-mpg"],df["price"],color="green")
plt.xticks(rotation=90)
plt.xlabel("highway-mpg",fontsize=15)
```

####  Observation
We know that car will attain max speed with respect to rpm, within(4500-5000rpm) minimum peak-rpm car attaining max speed are usualy costly(also notice the number of cylinders that are high) and from fig(2 and 3) we can absorb both are city and highway mpg(miles per gallon) are Negetively correlated, and contributing to price Equally

```python
sns.displot(x=df['horsepower'],y=df['price'],hue=df['body-style'])
```

#### Observation
Based on the horsepower comparing the price of cars,its linearly related to price of a automobile.Beacuse the horsepower which determines the top speed of the car and these super cars are taged as premium variant in the market.Its also distinguished that most of the premuim cars  are in 'Sedan ' Body style.
